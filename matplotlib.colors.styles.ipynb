```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,11,10)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x,color='purple')
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x,color='#bd1c2c')#hex color picker
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x,color='#bd1c2c', label='x vs x')
ax.plot(x,x+1,color="purple", label='x vs x+1')
ax.legend()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x,color='#bd1c2c',lw=10)
ax.plot(x,x+1,color="purple", linewidth=0.5)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x,color='#bd1c2c',lw=10, ls='--')#linestyle, '--', '-', '-.', ':', 'steps'
ax.plot(x,x+1,color="purple", linewidth=2, linestyle='-.')
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
lines=ax.plot(x,x+1,color="purple", linewidth=2)
lines[0].set_dashes([1,1,2,3,5])
```
